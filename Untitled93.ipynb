```python
import numpy as np
import matplotlib.pyplot as plt

# Constants from your model
alpha = 0.01  # Sensitivity of green investment
Ceh, Cde, Cet, Cep = 2, 3, 2, 2  # Carbon tax components
T = 360  # Time horizon
G_investment = 29.9  # Fixed green technology investment

# Range of demand (D)
demand_values = np.linspace(300, 1000, 100)  # Increasing demand

# Carbon emissions without green investment
def carbon_emissions_without_green(demand):
    CTH = Ceh * T * demand  # Carbon emissions in holding
    CTD = Cde * T * demand  # Carbon emissions in deterioration
    CTT = Cet * T * demand  # Carbon emissions in transportation
    CTP = Cep * T * demand  # Carbon emissions in production
    total_carbon = CTH + CTD + CTT + CTP
    return total_carbon

# Carbon emissions with green investment
def carbon_emissions_with_green(demand, G):
    CTH = Ceh * T * demand
    CTD = Cde * T * demand
    CTT = Cet * T * demand
    CTP = Cep * T * demand
    total_carbon = (CTH + CTD + CTT + CTP) * np.exp(-alpha * G)  # Green investment reduces emissions
    return total_carbon

# Calculate emissions for each demand value
carbon_emissions_no_green = carbon_emissions_without_green(demand_values)
carbon_emissions_with_green = carbon_emissions_with_green(demand_values, G_investment)

# Plot the results
plt.figure()

# Plot without green investment
plt.plot(demand_values, carbon_emissions_no_green, label="Without Green Investment", linestyle='--', color='red')

# Plot with green investment
plt.plot(demand_values, carbon_emissions_with_green, label=f'With Green Investment', color='green')


plt.xlabel('Demand')
plt.ylabel('Total Carbon Emissions')
plt.grid(True)
plt.legend()
plt.show()
```
