```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for investment in preservation technology
theta_0 = 0.01  # Base deterioration rate
omega = 0.05  # Preservation cost sensitivity
a1 = 0.03  # Sensitivity of preservation technology
xi = 20.1  # Investment in preservation technology (from optimal solutions)

# Compute the reduced deterioration rate and preservation effect
theta = theta_0 * np.exp(-omega * xi)
m_xi = (a1 * xi) / (1 + a1 * xi)

# Parameters from the numerical example
a = 500  # Initial demand rate (L/day)
b = 20   # Rate of increase in demand (L/day^2)
c = 400  # Constant demand during second phase (L/day)
e1 = 10  # Rate of decrease in demand (L/day^2)
Dlow = 300  # Low demand (L/day)
T = 360  # Total time horizon (days)
P = 984  # Production rate set to 984 L/day

# Phase durations
gamma1 = 90  # End of phase 1 (days)
gamma2 = 150  # End of phase 2 (days)
gamma3 = 240  # End of phase 3 (days)

# Time intervals for each phase
t1 = np.linspace(0, gamma1, 100)  # Phase 1: 0 to γ1
t2 = np.linspace(gamma1, gamma2, 100)  # Phase 2: γ1 to γ2
t3 = np.linspace(gamma2, gamma3, 100)  # Phase 3: γ2 to γ3
t4 = np.linspace(gamma3, T, 100)  # Phase 4: γ3 to T

# Phase I Inventory Level
I1 = (P / (theta * (1 - m_xi))) - ((a + b * t1) / (theta * (1 - m_xi))) + (b / (theta * (1 - m_xi))**2)

# Phase II Inventory Level
I1_end = I1[-1]  # Ending inventory at the end of Phase I
I2 = ((P - c) / (theta * (1 - m_xi))) * (1 - np.exp(-theta * (1 - m_xi) * (t2 - gamma1))) + I1_end * np.exp(-theta * (1 - m_xi) * (t2 - gamma1))

# Phase III Inventory Level
I2_end = I2[-1]  # Ending inventory at the end of Phase II
I3 = ((P - c) / (theta * (1 - m_xi))) + (e1 * (t3 - gamma2)) / (theta * (1 - m_xi)) - (e1 / (theta * (1 - m_xi))**2) + I2_end * np.exp(-theta * (1 - m_xi) * (t3 - gamma2))

# Phase IV Inventory Level
I3_end = I3[-1]  # Ending inventory at the end of Phase III
I4 = ((P - Dlow) / (theta * (1 - m_xi))) * (1 - np.exp(-theta * (1 - m_xi) * (t4 - gamma3))) + I3_end * np.exp(-theta * (1 - m_xi) * (t4 - gamma3))

# Concatenate all time and inventory values for a continuous graph
t_total = np.concatenate([t1, t2, t3, t4])
I_total = np.concatenate([I1, I2, I3, I4])

# Plot the inventory behavior graph
plt.figure(figsize=(10, 6))
plt.plot(t_total, I_total, label="Inventory Level", color='blue', linewidth=2)
plt.xlabel("Time (days)", fontsize=12)
plt.ylabel("Inventory Level (Liters)", fontsize=12)

# Custom x-axis labels using symbolic names
phase_labels = ['Phase 1 End', 'Phase 2 End', 'Phase 3 End']
phase_positions = [gamma1, gamma2, gamma3]

# Add vertical lines for phase transitions with labels
plt.axvline(x=gamma1, color='green', linestyle='--', label=phase_labels[0])
plt.axvline(x=gamma2, color='orange', linestyle='--', label=phase_labels[1])
plt.axvline(x=gamma3, color='red', linestyle='--', label=phase_labels[2])

# Set x-ticks to custom values (symbolic)
plt.xticks([0, gamma1, gamma2, gamma3, T], ['0', 'γ₁', 'γ₂', 'γ₃', 'T'])

plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
