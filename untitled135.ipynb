```python
import numpy as np
import matplotlib.pyplot as plt

# Adjusted Problem Constants for Model 2 (time in years)
S = 200    # Setup cost
h = 1.5    # Holding cost
Cd = 2     # Deterioration cost
Ct = 1     # Transportation cost
l = 25     # Fixed production cost
Cp = 40    # Production cost
Ceh = 2    # Carbon tax in holding
Cde = 3    # Carbon tax in deterioration
Cet = 2    # Carbon tax in transportation
Cep = 2    # Carbon tax in production
CT = 1     # Base carbon tax
alpha = 0.01  # Sensitivity of green investment
T = 360    # Time horizon
gamma1 = 90   # Time for phase 1
gamma2 = 150  # Time for phase 2
gamma3 = 240  # Time for phase 3
Dlow = 300    # Low demand
a = 500       # Initial demand coefficient (yearly)
b = 20        # Linear demand increase rate (yearly)
c = 400       # Constant demand in phase 2 (yearly)
e1 = 10       # Demand reduction rate in phase 3 (yearly)
omega = 0.05  # Preservation sensitivity
theta = 0.01  # Deterioration rate
a1 = 0.03     # Parameter for preservation effect
C_eco = 1.50  # Cost per unit of eco-friendly packaging
C_inc = 0.50 # Incentive or subsidy per unit

# PSO parameters for Model 2
num_particles = 30
iterations = 100
w_min, w_max = 0.4, 0.9  # Inertial weight range (for WPSO)
c1, c2 = 2.0, 2.0        # Cognitive and social coefficients
bounds_P = [40, 55]      # Adjusted bounds for P
bounds_xi = [20, 25]     # Adjusted bounds for xi
bounds_G = [20, 30]      # Adjusted bounds for green investment G

# Demand function phases (time in years)
def demand(t):
    if t <= gamma1:
        return a + b * t  # Yearly demand change
    elif t <= gamma2:
        return c
    elif t <= gamma3:
        return c - e1 * (t - gamma2)
    else:
        return Dlow

# Total demand integral over time (now in years)
def total_demand():
    return a * gamma1 + 0.5 * b * gamma1**2 + c * (gamma2 - gamma1) + \
           0.5 * e1 * (gamma3 - gamma2)**2 + Dlow * (T - gamma3)

# Inventory function based on deterioration rate
def inventory(P, xi):
    m_xi = a1 * xi / (1 + a1 * xi)  # Reduced deterioration rate
    I1 = (P - (a + b * gamma1)) / (theta * (1 - m_xi)) * (1 - np.exp(-theta * (1 - m_xi) * gamma1))
    I2 = (P - c) / (theta * (1 - m_xi)) * (1 - np.exp(-theta * (1 - m_xi) * (gamma2 - gamma1)))
    I3 = (P - c) / (theta * (1 - m_xi)) + e1 * (gamma3 - gamma2) / (theta * (1 - m_xi))
    I4 = (P - Dlow) / (theta * (1 - m_xi)) * (1 - np.exp(-theta * (1 - m_xi) * (T - gamma3)))
    return I1 + I2 + I3 + I4

# Carbon Taxes (time in years)
def carbon_taxes(P, xi, G):
    I_total = inventory(P, xi)
    CTH = Ceh * I_total
    CTD = Cde * theta * I_total
    CTT = Cet * total_demand()
    CTP = l * P * T + (Cep / P) * total_demand()
    CTReduced = CT * np.exp(-alpha * G) * (CTH + CTD + CTT + CTP)
    return CTReduced

# Eco-Friendly Packaging Cost
def eco_friendly_packaging_cost(G):
    return (C_eco - C_inc) * total_demand()

# Objective Function (Total Average Cost to Minimize)
def total_average_cost(P, xi, G):
    I_total = inventory(P, xi)

    # Calculate the different components of the cost function
    holding_cost = h * I_total
    deterioration_cost = Cd * theta * np.exp(-omega * xi) * I_total
    transportation_cost = Ct * total_demand()
    production_cost = l * P * T + (Cp / P) * total_demand()
    preservation_investment = xi
    green_investment = G

    # Carbon tax
    carbon_tax = carbon_taxes(P, xi, G)

    # Eco-friendly packaging cost
    eco_packaging_cost = eco_friendly_packaging_cost(G)

    # Penalty to adjust cost
    penalty = 0.01 * (P**2 + xi**2 + G**2)  # Non-linear penalty

    # Total cost
    total_cost = S + holding_cost + deterioration_cost + transportation_cost + production_cost + preservation_investment + carbon_tax + green_investment + eco_packaging_cost + penalty
    assert total_cost > 0, "Total cost must be positive"
    return total_cost / T  # Average cost per unit time

# WPSO implementation
def WPSO():
    # Initialize particles
    P_particles = np.random.uniform(bounds_P[0], bounds_P[1], num_particles)
    xi_particles = np.random.uniform(bounds_xi[0], bounds_xi[1], num_particles)
    G_particles = np.random.uniform(bounds_G[0], bounds_G[1], num_particles)
    velocities = np.zeros((num_particles, 3))  # Now with 3 decision variables
    P_best = P_particles.copy()
    xi_best = xi_particles.copy()
    G_best = G_particles.copy()

    # Initialize global best with the fitness of the first particle
    global_best = total_average_cost(P_particles[0], xi_particles[0], G_particles[0])
    global_best_position = (P_particles[0], xi_particles[0], G_particles[0])

    fitness_curve = []

    for iter in range(iterations):
        w = w_max - (w_max - w_min) * iter / iterations
        for i in range(num_particles):
            fitness = total_average_cost(P_particles[i], xi_particles[i], G_particles[i])
            if fitness < total_average_cost(P_best[i], xi_best[i], G_best[i]):
                P_best[i] = P_particles[i]
                xi_best[i] = xi_particles[i]
                G_best[i] = G_particles[i]
            if fitness < global_best:
                global_best = fitness
                global_best_position = (P_particles[i], xi_particles[i], G_particles[i])

        for i in range(num_particles):
            r1, r2 = np.random.rand(), np.random.rand()
            velocities[i][0] = (w * velocities[i][0] +
                                c1 * r1 * (P_best[i] - P_particles[i]) +
                                c2 * r2 * (global_best_position[0] - P_particles[i]))
            velocities[i][1] = (w * velocities[i][1] +
                                c1 * r1 * (xi_best[i] - xi_particles[i]) +
                                c2 * r2 * (global_best_position[1] - xi_particles[i]))
            velocities[i][2] = (w * velocities[i][2] +
                                c1 * r1 * (G_best[i] - G_particles[i]) +
                                c2 * r2 * (global_best_position[2] - G_particles[i]))
            P_particles[i] = np.clip(P_particles[i] + velocities[i][0], bounds_P[0] + np.random.uniform(0.1, 1), bounds_P[1] - np.random.uniform(0.1, 1))
            xi_particles[i] = np.clip(xi_particles[i] + velocities[i][1], bounds_xi[0] + np.random.uniform(0.1, 1), bounds_xi[1] - np.random.uniform(0.1, 1))
            G_particles[i] = np.clip(G_particles[i] + velocities[i][2], bounds_G[0] + np.random.uniform(0.1, 1), bounds_G[1] - np.random.uniform(0.1, 1))

        fitness_curve.append(global_best)

    return global_best, global_best_position, fitness_curve

# CPSO implementation
def CPSO():
    # Initialize particles
    P_particles = np.random.uniform(bounds_P[0], bounds_P[1], num_particles)
    xi_particles = np.random.uniform(bounds_xi[0], bounds_xi[1], num_particles)
    G_particles = np.random.uniform(bounds_G[0], bounds_G[1], num_particles)
    velocities = np.zeros((num_particles, 3))  # Now with 3 decision variables
    P_best = P_particles.copy()
    xi_best = xi_particles.copy()
    G_best = G_particles.copy()

    # Initialize global best with the fitness of the first particle
    global_best = total_average_cost(P_particles[0], xi_particles[0], G_particles[0])
    global_best_position = (P_particles[0], xi_particles[0], G_particles[0])

    fitness_curve = []

    phi = c1 + c2
    chi = 2 / (phi - 2 + np.sqrt(phi ** 2 - 4 * phi))  # Constriction factor

    for iter in range(iterations):
        for i in range(num_particles):
            fitness = total_average_cost(P_particles[i], xi_particles[i], G_particles[i])
            if fitness < total_average_cost(P_best[i], xi_best[i], G_best[i]):
                P_best[i] = P_particles[i]
                xi_best[i] = xi_particles[i]
                G_best[i] = G_particles[i]
            if fitness < global_best:
                global_best = fitness
                global_best_position = (P_particles[i], xi_particles[i], G_particles[i])

        for i in range(num_particles):
            r1, r2 = np.random.rand(), np.random.rand()
            velocities[i][0] = (chi * (velocities[i][0] +
                                       c1 * r1 * (P_best[i] - P_particles[i]) +
                                       c2 * r2 * (global_best_position[0] - P_particles[i])))

            velocities[i][1] = (chi * (velocities[i][1] +
                                       c1 * r1 * (xi_best[i] - xi_particles[i]) +
                                       c2 * r2 * (global_best_position[1] - xi_particles[i])))

            velocities[i][2] = (chi * (velocities[i][2] +
                                       c1 * r1 * (G_best[i] - G_particles[i]) +
                                       c2 * r2 * (global_best_position[2] - G_particles[i])))

            # Update particle positions and prevent extremes
            P_particles[i] = np.clip(P_particles[i] + velocities[i][0], bounds_P[0] + np.random.uniform(0.1, 1), bounds_P[1] - np.random.uniform(0.1, 1))
            xi_particles[i] = np.clip(xi_particles[i] + velocities[i][1], bounds_xi[0] + np.random.uniform(0.1, 1), bounds_xi[1] - np.random.uniform(0.1, 1))
            G_particles[i] = np.clip(G_particles[i] + velocities[i][2], bounds_G[0] + np.random.uniform(0.1, 1), bounds_G[1] - np.random.uniform(0.1, 1))

        fitness_curve.append(global_best)

    return global_best, global_best_position, fitness_curve

# Running the algorithms and plotting the convergence curve
best_wpso, best_wpso_position, wpso_curve = WPSO()
best_cpso, best_cpso_position, cpso_curve = CPSO()

# Plot convergence curve
plt.plot(wpso_curve, label='WPSO', linestyle='--', marker='o')
plt.plot(cpso_curve, label='CPSO', linestyle='-', marker='x')
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Best Fitness (Total Average Cost)')

plt.legend()
plt.show()

# Final outputs
print(f"Best WPSO solution: P = {best_wpso_position[0]}, xi = {best_wpso_position[1]}, G = {best_wpso_position[2]}, Cost = {best_wpso}")
print(f"Best CPSO solution: P = {best_cpso_position[0]}, xi = {best_cpso_position[1]}, G = {best_cpso_position[2]}, Cost = {best_cpso}")
```
