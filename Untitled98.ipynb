```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the parameters
parameters = ['Setup Cost (S)', 'Holding Cost (h)', 'Production Cost (Cp)',
              'Transportation Cost (Ct)', 'Preservation Sensitivity ()',
              'Carbon Tax in Holding (Ceh)', 'Carbon Tax in Transportation (Cet)',
              'Carbon Tax in Production (Cep)', 'Base Carbon Tax (CT)']

# Sensitivity analysis data for each parameter change (-20%, -10%, +10%, +20%)
cost_change_negative_20 = [3540.4530, 3712.6540, 3602.4055, 3503.8035, 3603.6550,
                           3615.1245, 3605.2005, 3603.2035, 3600.3055]
cost_change_negative_10 = [3592.5042, 3698.4545, 3627.9025, 3599.2025, 3632.4520,
                           3641.5545, 3630.0155, 3635.5525, 3633.5520]
cost_change_positive_10 = [3742.7828, 3612.1050, 3689.5020, 3740.6505, 3685.1035,
                           3694.8035, 3687.2505, 3687.2055, 3685.2525]
cost_change_positive_20 = [3801.5052, 3575.9020, 3713.4055, 3795.4015, 3715.5050,
                           3718.9555, 3714.1015, 3715.5025, 3715.5025]

# X-axis for plotting
x = np.arange(len(parameters))

# Plotting the data
plt.figure(figsize=(12, 7))
plt.plot(x, cost_change_negative_20, marker='o', label='-20% Change')
plt.plot(x, cost_change_negative_10, marker='o', label='-10% Change')
plt.plot(x, cost_change_positive_10, marker='o', label='+10% Change')
plt.plot(x, cost_change_positive_20, marker='o', label='+20% Change')

# Adding labels and title
plt.xticks(x, parameters, rotation=45, ha='right')
plt.xlabel('Parameters')
plt.ylabel('Total Cost')

plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
