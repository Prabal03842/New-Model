```python
import matplotlib.pyplot as plt
import numpy as np

# Model-specific parameters
T1, T2, T = 30, 60, 100  # Define time intervals
Dlow, Dpeak, Dmod = 100, 500, 300  # Demand levels in each phase
gamma1, gamma2, gamma3 = 0.01, 0.03, 0.015  # Adjusted growth rates

# Define time arrays for each phase
time_phase1 = np.linspace(0, T1, 100)
time_phase2 = np.linspace(T1, T2, 100)
time_phase3 = np.linspace(T2, T, 100)

# Demand functions for each phase
D_phase1 = Dlow * (1 - np.exp(-gamma1 * time_phase1))
D_phase2 = Dpeak * (1 - np.exp(-gamma2 * (time_phase2 - T1)))
D_phase3 = Dmod * (1 - np.exp(-gamma3 * (time_phase3 - T2)))

# Combine time and demand data for a continuous line
time_full = np.concatenate([time_phase1, time_phase2, time_phase3])
D_full = np.concatenate([D_phase1, D_phase2, D_phase3])

# Plotting the demand graph with phase markers
plt.figure(figsize=(10, 6))
plt.plot(time_full, D_full, label='Demand Rate', color='blue', linewidth=2)

# Adding phase markers
plt.axvline(x=T1, color='red', linestyle='--', label='End of Low Demand Phase')
plt.axvline(x=T2, color='green', linestyle='--', label='End of Peak Demand Phase')

# Shading for phases
plt.fill_betweenx([0, max(Dpeak, Dmod)], 0, T1, color='blue', alpha=0.1)
plt.fill_betweenx([0, max(Dpeak, Dmod)], T1, T2, color='orange', alpha=0.1)
plt.fill_betweenx([0, max(Dpeak, Dmod)], T2, T, color='green', alpha=0.1)

# Adding labels, title, and legend
plt.xlabel('Time (hours)')
plt.ylabel('Demand Rate (kg/hour)')
plt.title('Demand Rate vs Time with Phase Markers')
plt.legend()
plt.grid(True)
plt.show()
```
